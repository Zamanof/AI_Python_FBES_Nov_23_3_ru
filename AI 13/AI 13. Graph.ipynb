```python
import torch
```

```python
x = torch.tensor(2.0, requires_grad=True)
y = x**2 + 5*x + 7

print(x)
print(y)
print(y.grad_fn)
```

```python
x = torch.tensor(3.0, requires_grad=True)
a = x * 2
b = a + 4
c = b ** 2
y = c

print(a)
print(b)
print(c)
print(y)
print(y.grad_fn)
print(y.grad_fn.next_functions)
```

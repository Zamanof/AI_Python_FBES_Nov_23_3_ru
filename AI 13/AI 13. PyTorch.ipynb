```python
import torch
```

### Тензор
- Скаляр - одно число - 0 мерный тензор
- Вектор - список чисел - 1 мерный тензор
- Матрица - 2 мерный тензор
- Изображение(RGB) - 3 мерный тензор
- Пачка изображений - 4 мерный тензор

```python
scalar = torch.tensor(6)
print(f"Scalar: {scalar}, Shape: {scalar.shape}")

vector = torch.tensor([5, 9, 7])
print(f"Vector: {vector}, Shape: {vector.shape}")

matrix = torch.tensor([[25, 65],[12, 32],[33,33]])
print(f"Matrix: {matrix}, Shape: {matrix.shape}")

tensor3d = torch.randn((5, 5, 5))
print(f"Shape: {tensor3d.shape}")
```

```python
# tensor generate
print(torch.zeros(5, 5))
print(torch.ones(5, 5))
print(torch.rand(5, 5))
print(torch.arange(0, 10, 2))
```

```python
a =torch.tensor([1, 2, 3])
b = torch.tensor([4, 5, 6])
print(a + b) # Сложение
print(a * b) # Покомпонентное умножение
print(torch.dot(a, b)) # Скалярное произведение
```

```python
A = torch.tensor([[2, 3], [6, 4]], dtype=torch.float32)
B = torch.tensor([[5, 6], [7, 8]], dtype=torch.float32)

print(A @ B)
```

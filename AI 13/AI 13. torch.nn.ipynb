```python
linear = nn.Linear(in_features=3, out_features=2)

x = torch.tensor([1.0, 2.0, 3.0])
y = linear(x)
print(f"Vxod x: {x}")
print(f"Vixod y: {y}")
print(f"Forma vixoda: {y.shape}")

print(f"W: {linear.weight}")
print(f"b: {linear.bias}")
```

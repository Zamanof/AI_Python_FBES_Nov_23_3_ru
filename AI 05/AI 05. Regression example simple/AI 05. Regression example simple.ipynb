```python
from cProfile import label

import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([52, 57, 62, 70, 74])
```

```python
n = len(x)
```

```python
b = (n * np.sum(x*y) - np.sum(x) * np.sum(y)) / (n*np.sum(x**2) - (np.sum(x)**2))
a = (np.sum(y) - b*np.sum(x))/n
```

```python
print(f"Уравнение: y = {a:.2f} = {b:.2f}")
```

```python
# предсказание для графика
y_pred = a + b * x
```

```python
plt.scatter(x, y, color='red', label='Данные')
plt.plot(x, y, color="blue", label='Регрессия')
plt.xlabel('Часы подготовки')
plt.ylabel('Оценка')
plt.legend()
plt.title("Простая линейная регрессия")
plt.show()
```

```python
while True:
    try:
        x_new = float(input("Введите количество часов подготовки"))
        if x_new == 0:
            print("Exit")
            break
        y_new = a + b * x_new
        print(f"Прогнозируемая оценка: {y_new:.2f}")
        plt.scatter(x, y, color='red', label='Данные')
        plt.plot(x, y, color="blue", label='Регрессия')
        plt.scatter(x_new, y_new, color='green', s=100, label="Ваш прогноз", zorder=5)
        plt.xlabel('Часы подготовки')
        plt.ylabel('Оценка')
        plt.legend()
        plt.title(f"Прогноз для x = {x_new}")
        plt.show()
    except ValueError:
        print("Введите число")
```

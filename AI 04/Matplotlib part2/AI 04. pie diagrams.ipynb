```python
import matplotlib.pyplot as plt
```

```python
labels = ["Baku", "Sumqayit", "Ganja"]
percents = [50, 35, 15]
```

```python
plt.pie(percents, labels=labels)
plt.title("Pie diagram")
plt.show()
```

```python
plt.pie(percents, labels=labels, autopct="%1.1f%%")
plt.title("Pie diagram")
plt.show()
```

```python
explode = (0, 0.1, 0)
plt.pie(percents, labels=labels, explode=explode, autopct="%1.1f%%", startangle=90, shadow=True,
        colors=["red", "green", "blue"])
plt.title("Pie diagram")
plt.show()
```

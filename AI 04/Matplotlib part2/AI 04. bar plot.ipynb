```python
import matplotlib.pyplot as plt
```

```python
languages = ["English", "Chinese", "Azerbaijani", "Russian"]
words = [600_000, 370_000, 100_000, 250_000]
```

```python
plt.bar(languages, words, color="#008000")
plt.xlabel("Languages")
plt.ylabel("Words")
plt.title("World languages words diagram")
plt.show()
```

```python
plt.barh(languages, words, color=["#008000", "red", "blue", "orange"])
plt.xlabel("Languages")
plt.ylabel("Words")
plt.title("World languages words diagram")
for i , value in enumerate(words):
    plt.text(i, i+0.4, str(value), ha="left")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,10)
```

```python
y = 2*x
```

```python
fig = plt.figure()
axes = fig.add_axes((0, 0, 1, 1))
axes.plot(x,y);
```

```python
a = np.linspace(0, 10, 11)
b = a**4
```

```python
a
```

```python
b
```

```python
# fig = plt.figure(dpi=100, figsize=(5, 8))
fig = plt.figure()
axes = fig.add_axes((0, 0, 1, 1))
axes.plot(a,b);

axes2 = fig.add_axes((0.5, 0.5, 0.3, 0.3))
axes2.plot(x,y)
axes.set_xlim(0)
axes.set_ylim(0)
axes2.set_ylim(0)
axes2.set_xlim(0)
axes.set_xlabel('a')
axes.set_ylabel('b')
axes.set_title('Axes Title')
axes2.set_title('Small Axes Title')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y)
axes[1].plot(a,b)
plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
axes[0, 0].plot(x,y)
axes[1, 1].plot(a,b)
plt.tight_layout()
# fig.subplots_adjust(wspace=0.5, hspace=0.1)
```

```python
fig = plt.figure()
axes = fig.add_axes((0, 0, 1, 1))
axes.plot(x,y, label='y = 2x', color='#e00dcb')
axes.plot(x,x**2, label='y = x**2', color='blue')
axes.plot(x,x + 2, label='y = x+2', color='red', lw=8)
# axes.legend(loc='upper left')
axes.legend(loc=(1, 0))
plt.show()
```

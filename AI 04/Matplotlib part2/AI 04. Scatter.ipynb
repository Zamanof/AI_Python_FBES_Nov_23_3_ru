```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [4, 6, 8, 12, 14, 16]
y = [45, 3, 23, 5, 49, 50]
```

```python
plt.scatter(x, y)
plt.show()
```

```python
x = np.random.random(100)
y = np.random.random(100)
colors = np.random.random(100)
sizes = 500 * np.random.random(100)
plt.scatter(x, y, c=colors, s=sizes, alpha=0.8, cmap="ocean")
plt.show()
plt.show()
```

```python
plt.scatter(x, y, color="red", label="red", marker="x")
plt.scatter(y, x, color="blue", label="blue", marker="s")
plt.legend()
plt.show()
```
